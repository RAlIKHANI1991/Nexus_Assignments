## 1. Log and Exp: Proving Inverse Relationship


```python
import numpy as np

# Show log(exp(x)) == x and exp(log(x)) == x for several values
x_vals = np.array([1 , 2 , 3 , np.e])

print("x\tlog(exp(x))\texp(log(x))")
for x in x_vals:
    print(f"{x:.2f}\t{np.log(np.exp(x)):.2f}\t\t{np.exp(np.log(x)):.2f}")

# Edge case: try negative and zero values (show error/NaN for log)
y_vals = np.array([0, -1, 5])
print("\nlog(exp(y)) for 0 and negative values:")
print(np.log(np.exp(y_vals)))
```

## 4. Features of the Sigmoid Function

```python
import sympy as sym

x = sym.Symbol('x')  # ÿ™ÿπÿ±€åŸÅ ŸÖÿ™ÿ∫€åÿ± ŸÜŸÖÿßÿØ€åŸÜ x
sig = 1 / (1 + sym.exp(-x))  

sym.pprint(sig)  
```

```python

def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x = np.linspace(-10, 10, 200)  # 200 num geberate between -10 , 10 
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("œÉ(x)")
plt.grid(True)
plt.show()

print("Range:", (y.min(), y.max()), "| At x=0:", sigmoid(0))
```

## 5. Derivatives of Famous Functions

```python
x = np.linspace(-3, 3, 400)   # 400 number generate between -3 , 3 

plt.plot(x, x**2, label="x¬≤")
plt.plot(x, 2*x , '--', label="d/dx x¬≤")

x = np.linspace(0, 3, 400)
plt.plot(x, np.exp(x), label="exp(x)")
plt.plot(x, np.exp(x), '--', label="d/dx exp(x)")

plt.plot(x, np.log(x + 0.1), label="log(x)")
plt.plot(x, 1 / (x + 0.1) , '--', label="d/dx log(x)")

plt.legend()
plt.title("Functions and Their Derivatives")
plt.show()
```

```python
# more example
import matplotlib.pyplot as plt

x = np.linspace(0.1, 3, 400)
y = np.log(x + 0.1)

dy = np.diff(y)
dx = np.diff(x)
derivative = dy / dx

x_mid = (x[:-1] + x[1:]) / 2  # ŸÖ€åÿßŸÜ⁄Ø€åŸÜ ŸÜŸÇÿßÿ∑ ÿ®ÿ±ÿß€å ŸÖÿ≠Ÿàÿ± x ŸÖÿ¥ÿ™ŸÇ

plt.plot(x, y, label="log(x + 0.1)")
plt.plot(x_mid, derivative, '--', label="Approx. Derivative")
plt.legend()
plt.grid(True)
plt.title("Function and Its Approximate Derivative")
plt.show()
```

## 6. Gradient of Selected Functions

```python
# search and learn more about sumpy
import sympy as sp

# Define symbolic variables
x, y = sp.symbols('x y')

# Define the function
f = x**2 * y + sp.sin(y)

# Compute partial derivatives
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("Function: f(x, y) =", f)
print("Partial derivative w.r.t x (‚àÇf/‚àÇx):", df_dx)
print("Partial derivative w.r.t y (‚àÇf/‚àÇy):", df_dy)
```
