### Task 1 ‚Äî Mock temperatures



```python
# üëâ # TODO: import numpy and create an array of 365
# normally-distributed ¬∞C values (¬µ=20, œÉ=5) called temps

import numpy as np
import matplotlib.pyplot as plt 
temps = np.random.normal(loc=20 , scale=5 , size=365)

# ÿ±ÿ≥ŸÖ Ÿá€åÿ≥ÿ™Ÿà⁄Øÿ±ÿßŸÖ
plt.hist(temps, bins=30 , edgecolor='black', density=True)

# ÿ±ÿ≥ŸÖ ŸÖŸÜÿ≠ŸÜ€å ⁄Ü⁄ØÿßŸÑ€å ÿßÿ≠ÿ™ŸÖÿßŸÑ ÿ±Ÿà€å Ÿá€åÿ≥ÿ™Ÿà⁄Øÿ±ÿßŸÖ
from scipy.stats import norm
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)

p = norm.pdf(x, np.mean(temps), np.std(temps)) #pdf
plt.plot(x, p, 'r', linewidth=2)

# ÿ™ŸÜÿ∏€åŸÖÿßÿ™ ŸÜŸÖŸàÿØÿßÿ±
plt.title("Normal distribution yearly")
plt.xlabel("temp of daily")
plt.ylabel("Probability density")

plt.grid(True)
plt.show()
```

### Task 2 ‚Äî Average temperature


```python
# üëâ # TODO: print the mean of temps
temps_mean = round((np.mean(temps)) , 2 )

print(temps_mean.dtype)
print(temps_mean)
```

### Task 3 ‚Äî Load ride log


```python
# üëâ # TODO: read "rides.csv" into df
# (columns: date,temp,rides,weekday)
import pandas as pd 
df = pd.read_csv("rides.csv")
df.head(7)
```

### Task 4 ‚Äî Weekday averages

```python
# üëâ # TODO: compute and print mean rides per weekday
df_mean_weekday = df.groupby('weekday')[['rides']].mean().reset_index()
df_mean_weekday
```

### Task 5 ‚Äî Scatter plot

```python
# üëâ # TODO: scatter-plot temperature (x) vs rides (y) from df
import matplotlib.pyplot as plt 

x = df['temp']
y = df['rides']

plt.scatter(x, y)  

plt.xlabel('Temperature (¬∞C)')
plt.ylabel('Number of Rides')
plt.title('Temperature vs. Rides')
plt.grid(True)

plt.show()

```
